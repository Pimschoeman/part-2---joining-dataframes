Importing the pandas library for creating the dataframe

```python
import pandas as pd
```

reading the wikipedia table into a dataframe

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df.head(10)
```

creating a new dataframe which excludes the rows where borough == "not assigned"

```python
# i assumed that i could make a new dataframe with specifying that "not assigned" would not appear in column borough. 
dfwna= df[df.Borough != 'Not assigned']
```

```python
dfwna.head(10)
```

Modifying the dataframe so values would be combined

```python
dfwna =dfwna.groupby(['Postcode','Borough'])
dfwna=dfwna['Neighbourhood'].apply(', '.join).reset_index() 
dfwna.head(10)
```

Iterating through the dataframe and assigning the borough value to the neighbourhood value

```python
for i in range(len(dfwna)):
    if dfwna['Neighbourhood'].iloc[i] =='Not assigned':
        dfwna['Neighbourhood'].iloc[i]= dfwna['Borough'].iloc[i]
        
dfwna
```

```python
#to be sure, again importing pandas.
#after importing pandas, appending the dataframe to variable dfpostal
#then printing the head of the dfpostal datafram
import pandas as pd
dfpostal = pd.read_csv('postaldata.csv')
dfpostal.head()
dfpostal.shape
```

```python
#In this step i am merging the two dataframes
#setting the variables df1 and df2
#naming the new dataframe df3
#joining on the postal value
df1=dfwna # first dataframe  
#renaming the column name in the first dataframe so it matches with the second
df1 = df1.rename(columns={'Postcode': 'Postal Code'})
df2=dfpostal #second dataframe
df3 = pd.merge(df1,df2,on=['Postal Code']) #merging the dataframes based on postalcode
df3 #seeing the new dataframe
```
